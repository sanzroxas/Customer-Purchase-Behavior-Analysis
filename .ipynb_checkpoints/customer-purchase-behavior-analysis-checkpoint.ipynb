```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import openpyxl
```

```python
df = pd.read_excel('Online Retail.xlsx')
```

```python
print(df.head())
print(df.info())
print(df.describe())
print(df.shape)
```

```python
print(df.isnull().sum())
# Example: Drop rows with missing CustomerID
df = df.dropna(subset=['Customer ID'])
```

```python
df = df.dropna(subset=['Customer ID']).copy()
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Customer ID'] = df['Customer ID'].astype(str)
```

```python
print(df.duplicated().sum())
df = df.drop_duplicates()
```

```python
sns.boxplot(x=df['Quantity'])
plt.show()

# Example: Remove rows where Quantity is less than 0 (likely returns)
df = df[df['Quantity'] >= 0]
```

```python
df['TotalPurchase'] = df['Quantity'] * df['Price']
```

```python
# Resample to daily, weekly, or monthly
daily_sales = df.resample('D', on='InvoiceDate')['TotalPurchase'].sum()
weekly_sales = df.resample('W', on='InvoiceDate')['TotalPurchase'].sum()
monthly_sales = df.resample('ME', on='InvoiceDate')['TotalPurchase'].sum()

# Plot
plt.figure(figsize=(12, 6))
monthly_sales.plot()
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

The monthly sales trend shows a general increase in sales from January to November. However, there is a very sharp decline in sales in December. This is most likely due to the dataset only containing data for the first part of December. Further investigation is needed to confirm this, but the data appears incomplete. Overall, the data shows strong sales performance throughout the year leading up to the holiday period

```python
# Find the most recent date in the dataset
most_recent_date = df['InvoiceDate'].max()

# Calculate Recency, Frequency, and Monetary Value
rfm = df.groupby('Customer ID').agg({
    'InvoiceDate': lambda x: (most_recent_date - x.max()).days,  # Recency
    'Invoice': 'nunique',  # Frequency
    'TotalPurchase': 'sum'  # Monetary Value
})

# Rename columns
rfm.columns = ['Recency', 'Frequency', 'MonetaryValue']

print(rfm.head())
```

```python
# Create quantiles (quintiles in this case)
r_labels = range(5, 0, -1)  # Recency scores (5 is best, 1 is worst)
f_labels = range(1, 6)  # Frequency scores (1 is best, 5 is worst)
m_labels = range(1, 6)  # MonetaryValue scores (1 is best, 5 is worst)

rfm['R_Score'] = pd.qcut(rfm['Recency'], q=5, labels=r_labels)
# Examine the distribution of Frequency (important!)
print(rfm['Frequency'].value_counts().sort_index())

# Define bin edges manually (example - adjust these!)
bins = [0, 1, 2, 3, 5, 205]  # Example: 0-1, 1-2, 2-3, 3-5, 5+ purchases

rfm['F_Score'] = pd.cut(rfm['Frequency'], bins=bins, labels=f_labels, include_lowest=True)
rfm['M_Score'] = pd.qcut(rfm['MonetaryValue'], q=5, labels=m_labels)

print(rfm.head())
```

```python
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
print(rfm.head())
```

```python
seg_map = {
    r'[1-2][1-5][1-5]': 'Hibernating',  # Low Recency = Inactive
    r'[3-4][1-5][1-5]': 'Active',      # Medium Recency = Engaged
    r'5[1-2][1-5]': 'New Customers', # High Recency, Low Freq = New
    r'5[3-5][1-2]': 'Promising',    # High Recency, High Freq, Low Spend = Growing
    r'5[3-5][3-5]': 'Champions'     # High Recency, High Freq, High Spend = Best
}
rfm['Segment'] = rfm['RFM_Score'].replace(seg_map, regex=True)
```

```python
  # Count the number of customers in each segment
segment_counts = rfm['Segment'].value_counts().sort_values(ascending=False)

# Create a bar plot
fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(x=segment_counts.index, y=segment_counts.values, ax=ax)
ax.set_title('Customer Segments')
ax.set_xlabel('Segment')
ax.set_ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import squarify  

segment_counts = rfm['Segment'].value_counts()
plt.figure(figsize=(12, 8))
squarify.plot(sizes=segment_counts, label=segment_counts.index, alpha=.8, color=sns.color_palette("viridis", len(segment_counts)))
plt.axis('off')
plt.title('Customer Segments (Treemap)')
plt.show()
```
